```python
new_book=book.drop_duplicates(keep='first')
new_book
```

```python
user_book1 = new_book.pivot(index='userid',columns='booktitle',values='bookrating').reset_index(drop=True)
user_book1
```

```python
user_book1.index = new_book.userid.unique()
user_book1
```

```python
#Impute those NaNs with 0 values
user_book1.fillna(0, inplace=True)
user_book1
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book1.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index =new_book.userid.unique()
user_sim_df.columns =new_book.userid.unique()
```
